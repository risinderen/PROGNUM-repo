```python
#3.2
from scipy.constants import G
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
#Sets the range to be the length of the list
m = range(len(masses))        
r = range(len(radii))

v_esc_list = []

for i in m:
    v_esc = ((2*G*masses[i])/radii[i])**0.5
    print(f"For {names[i]} with a mass of {masses[i]:.2e} kg, the speed is:")
    print(f"{v_esc:.3e} m/s")
    v_esc_list.append(v_esc)
    
v_esc_dictionary = dict(zip(names, v_esc_list))

print(v_esc_dictionary)
```

```python
#3.7
import numpy as np

def angle(d,D):        #define angle as the function of d and D
    a = 2*np.arctan(d/(2*D))        #Calculates the angle in radians
    degrees = np.degrees(a)        #Turns the angle into degrees
    return degrees

d = eval(input("enter the diameter of the object in km: "))
D = eval(input("enter the distance to the object in km: "))

print(angle(d,D))
```

```python
#3.10
import numpy as np

def cylinder(r,h):        #definining cylinder as the function of r and h
    
    A = 2*np.pi*r**2 + 2*np.pi*r*h       #the area
    V = np.pi*r**2*h       #Volume.
    return A, V

area, volume = cylinder(r,h)      #changes A to an output of area and V to an output of volume to use outside of the defined function.

r = eval(input("enter the radius of the cylinder: "))
h = eval(input("enter the height of the cylinder: "))

dr_Ratio = area/volume
print(cylinder(r,h))
print(dr_Ratio)
```

```python
import numpy as np

#3.17
N = eval(input("input the amount of digits: "))
randomraaa = np.random.normal(0, 2, N)
def stdev(N,randomraaa):
    miu_sum = 0
    s = 0
    
    for i in range(0, N):
        
        miu_sum += randomraaa[i]       #sums up the values of of the number generator
    
    miu = miu_sum/N       #deviding by the amount of numbers
    
    for i in range(0, N):
        
        s += (randomraaa[i]-miu)**2       #summing up the values of the number generator minus the average squared
    
    sigma = (s/N)**0.5
    
    return sigma

print(stdev(N,randomraaa))
print(np.std(randomraaa))
```
