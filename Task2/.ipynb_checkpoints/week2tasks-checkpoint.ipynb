```python
#2.21
A = 1
B = 2
i = 0
while i < 98:
    C = A + B     # Calculating the next value in the fibonacci sequence
    A = B         # Changing the value n-1 to n-2
    B = C         # Changing the value of n to n-1
    i = i + 1        # Adding 1 to the loop
print(C)
            
```

```python
#2.19
import numpy as np
a = np.random.normal(loc=0.0, scale=1.4, size=100000)        # Random number generator with center at 0, with a standard deviation of 1.4, with 100000 numbers.
b = np.std(a)        # Calculates the standard deviation of a.

print(a)
print(b)
```
