```python
#2.21
A = 1
B = 2
i = 0
while i < 98:
    C = A + B     #calculating the next value in the fibonacci sequence
    A = B         #changing the value n-1 to n-2
    B = C         #Changing the value of n to n-1
    i = i + 1        #Adding 1 to the loop
print(C)
            
```
