```python
#4.4
import numpy as np
import matplotlib.pyplot as plt
import scipy as scp
boop = eval(input("enter a value: "))

human_length = np.random.normal(loc=185, scale=8, size=100000)        #mean at 185, std of 8, 100000 samples

npmean = np.mean(human_length)          #mean
npstd = np.std(human_length)            #standard deviation

plt.hist(human_length, bins=40, )         #makes a histogram with the values of human_length with a total of 40 bars
lowerbound = npmean-npstd
upperbound = npmean+npstd

plt.xlabel("length in cm")        
plt.ylabel("amount of people")

plt.axvline(x=npmean, color='red', label=f"mean")
plt.axvspan(lowerbound, upperbound, color='green',zorder=-1)


plt.legend(title=f"the mean is {npmean:.2f} cm \nthe standard deviation is {npstd:.2f} cm")       #The legend. (legenda)
plt.show()
prob = scp.stats.norm.cdf(boop, loc = npmean, scale=npstd) * 100         #checks what the probability is of boop being smaller than the mean and the std
print(f" the probability of the height being smaller than the mean: {prob}")
prob2 = scp.stats.norm.cdf(lowerbound, loc = npmean, scale=npstd)* 100        #the probability of the lowerbound (mean minus std)
prob3 = scp.stats.norm.cdf(upperbound, loc = npmean, scale=npstd)* 100        #the probability of the upperbound (mean plus std)
probinbetween = abs(prob3-prob2)

print(f" the probability of the height being being between the lowerbound and upperbound: {probinbetween}")

```

```python
#4.12
import numpy as np
from matplotlib import pyplot as plt

x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

angle_degs = 30
angle_rads = np.radians(angle_degs)
Rotor = np.cos(angle_rads)+1j*np.sin(angle_rads)       #makes the array into the rotor

z = x + 1j*y      #makes it imaginary
zr = z*Rotor      #rotates
xr = zr.real      #Takes the lines for the x axis
yr = zr.imag      #Takes the lines for the y axis


plt.plot(x,y)
plt.plot(xr,yr, color='green')
plt.xlim([-1, 5])
plt.ylim([-1, 5])
plt.gca().set_aspect(1)
plt.show
```

```python
#4.21
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

x = np.random.normal(loc=0, scale=1, size= 100000)
y = np.random.normal(loc=2, scale=1, size= 100000)

xmean = np.mean(x)
xstd = np.std(x)
ymean = np.mean(y)
ystd = np.std(y)

plt.hist(x, bins=30, density=True, color='purple', edgecolor='black', alpha=0.7, label="X")
plt.hist(y, bins=30, density=True, color='red', edgecolor='black', alpha=0.7, label="Y")


x_vals = np.linspace(min(x.min(), y.min()), max(x.max(), y.max()), 1000)

x2 = norm.pdf(x_vals, scale=xstd, loc=xmean)
y2 = norm.pdf(x_vals, scale=ystd, loc=ymean)


plt.plot(x_vals, x2, color='black', linestyle=':', linewidth=5, label="X(PDF)")
plt.plot(x_vals, y2, color='black', linestyle='--', linewidth=5, label="Y(PDF)")

plt.grid(axis='y', linestyle='--')

plt.annotate("",
             xy=(-2, 0.15),        # arrow end
             xytext=(-3, 0.15),    # arrow start
             arrowprops=dict(arrowstyle="-|>", linewidth=3))

plt.annotate("",
             xy=(4, 0.15),        # arrow end
             xytext=(5, 0.15),    # arrow start
             arrowprops=dict(arrowstyle="-|>", linewidth=3))
plt.text(4.5, 0.16,
        r'$\mathcal{N}(2,1)$',
        fontsize = 14)
plt.text(-4.5, 0.16,
        r'$\mathcal{N}(0,1)$',
        fontsize = 14)
plt.legend(loc="upper right")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(0)
X = np.random.normal(0,1,100000)
Y = np.random.normal(2,1,100000)

# 2D histogram
H, xedges, yedges = np.histogram2d(X, Y, bins=30)

# Normalize to probability
H = H / H.sum()

# Flatten and sort bins by probability
Hflat = H.flatten()
Hsorted = np.sort(Hflat)[::-1]

# Cumulative probability
cdf = np.cumsum(Hsorted)
print(cdf)
# Sigma probability levels
levels_prob = [0.68, 0.95, 0.997]

# Find density thresholds
levels = []
for p in levels_prob:
    levels.append(Hsorted[np.searchsorted(cdf, p)])
levels = np.sort(levels) 

# Plot histogram
plt.figure(figsize=(8,6))
h = plt.hist2d(X, Y, bins=30, cmap="Blues")

# Compute bin centers
xcenters = 0.5*(xedges[1:] + xedges[:-1])
ycenters = 0.5*(yedges[1:] + yedges[:-1])

# Contours from dataset counts
c = plt.contour(xcenters, ycenters, H.T, levels=levels, colors='red')

# Label contours
plt.clabel(c, fmt={levels[0]:'1σ', levels[1]:'2σ', levels[2]:'3σ'})

plt.xlabel("X")
plt.ylabel("Y")
plt.title("2D Histogram with σ Contours from Dataset")

plt.grid(True, linestyle='--', alpha=0.5)
plt.colorbar(h[3], label="Counts")
plt.show()
```
